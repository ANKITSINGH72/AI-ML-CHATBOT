**Introduction**

This AI-powered financial chatbot is designed to respond to predefined financial queries using data extracted from the 10-K filings of Microsoft, Tesla, and Apple for the last three fiscal years. The chatbot provides quick and accurate responses to questions regarding key financial metrics such as total revenue, net income changes, total assets, total liabilities, and cash flow from operating activities.



**How It Works:**

The chatbot works by normalizing user queries (converting to lowercase and removing extra spaces) and matching these queries against a predefined set of financial questions. The chatbot uses simple string matching techniques to determine the appropriate financial data to return based on the user's query.

# Query Processing:

**Normalization:** User input is converted to lowercase and stripped of leading or trailing spaces to ensure consistency in query processing.

**Matching:** The chatbot looks for key phrases in the query (e.g., "total revenue," "net income," "total assets") and matches them to predefined responses.

**Data Retrieval:** The chatbot retrieves the relevant financial data from a preloaded dataset and returns it as the response.

**Supported Queries**

The chatbot can handle the following predefined queries:


**1-Total Revenue:** **bold text**
Query Example: "What is the total revenue?"

Response: The chatbot will return the total revenue for the most recent fiscal year for Microsoft, Tesla, and Apple.

**2- Net Income Change:**
Query Example: "How has net income changed over the last year?"

Response: The chatbot will return the percentage change in net income from the previous fiscal year for each company.

**3- Total Assets:**
Query Example: "What are the total assets?"

Response: The chatbot will return the total assets for the most recent fiscal year for each company.

**4- Total Liabilities:**
Query Example: "What are the total liabilities?"

Response: The chatbot will return the total liabilities for the most recent fiscal year for each company.

**5- Cash Flow from Operating Activities:**
Query Example: "What is the cash flow from operating activities?"

Response: The chatbot will return the cash flow from operating activities for the most recent fiscal year for each company.

```python
import pandas as pd

# Sample financial data for Microsoft, Tesla, and Apple (from Task 1)
data = {
    'Company': ['Microsoft', 'Microsoft', 'Microsoft', 'Tesla', 'Tesla', 'Tesla', 'Apple', 'Apple', 'Apple'],
    'Year': [2021, 2022, 2023, 2021, 2022, 2023, 2021, 2022, 2023],
    'Total Revenue ($M)': [168088, 198270, 211915, 53823, 81462, 105225, 365817, 394328, 408732],
    'Net Income ($M)': [61271, 72738, 74993, 5519, 12556, 14117, 94680, 99803, 100562],
    'Total Assets ($M)': [333779, 364840, 403382, 62131, 82594, 93000, 351002, 352755, 379000],
    'Total Liabilities ($M)': [191791, 198320, 215637, 28356, 30551, 35000, 287912, 302083, 310000],
    'Cash Flow from Operating Activities ($M)': [76742, 81758, 89002, 11500, 14709, 17800, 104038, 110174, 112000]
}

df = pd.DataFrame(data)

# Function to handle queries
def simple_chatbot(user_query):
    # Normalize the query: convert to lowercase and strip leading/trailing spaces
    user_query = user_query.strip().lower()

    # Check for various financial queries using flexible matching
    if "total revenue" in user_query:
        response = df.groupby('Company')['Total Revenue ($M)'].last().to_string()
    elif "net income" in user_query and "change" in user_query:
        df['Net Income Change (%)'] = df.groupby('Company')['Net Income ($M)'].pct_change() * 100
        response = df.groupby('Company')['Net Income Change (%)'].last().to_string()
    elif "total assets" in user_query:
        response = df.groupby('Company')['Total Assets ($M)'].last().to_string()
    elif "total liabilities" in user_query:
        response = df.groupby('Company')['Total Liabilities ($M)'].last().to_string()
    elif "cash flow" in user_query or "operating activities" in user_query:
        response = df.groupby('Company')['Cash Flow from Operating Activities ($M)'].last().to_string()
    else:
        response = "Sorry, I can only provide information on predefined queries."

    return response

# Example interaction
user_query = input("Ask a financial question: ")
print(simple_chatbot(user_query))
```
